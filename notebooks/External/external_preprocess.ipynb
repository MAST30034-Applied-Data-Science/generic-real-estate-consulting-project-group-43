Population_sa2 preprocess

```python
xls = pd.ExcelFile('../../data/raw/external_dataset/population_sa2.csv')
sheets = xls.sheet_names
sheets
```

```python
population_sa2 = pd.read_excel('../../data/raw/external_dataset/population_sa2.csv', sheet_name=['Table 1'], header =  None)['Table 1'].drop(range(6))
year = population_sa2.iloc[0].dropna()
header = population_sa2.iloc[1]
header.iloc[10:].update(year)
population_sa2 = population_sa2[3:]
population_sa2.columns = header 
population_sa2 = population_sa2.drop(columns=['2011-2021', 'Area', '%'])
population_sa2
```

```python
vic_population_sa2 = population_sa2[population_sa2['S/T name'] == 'Victoria']
vic_population_sa2 = vic_population_sa2.dropna(axis='columns')
vic_population_sa2 = vic_population_sa2.drop(columns = ['SA4 code', 'SA4 name', 'SA3 code', 'SA3 name', 'S/T code', 'GCCSA code', 'GCCSA name'])
vic_population_sa2
```

```python
vic_population_sa2.to_csv('../../data/curated/vic_population_sa2.csv')
```

```python
def vic_income(file):
    income = pd.read_excel(f'../../data/raw/external_dataset/{file}.csv', sheet_name=['Table 1.4'], header =  None)['Table 1.4'].drop(range(6))
    income = income.iloc[:, :7]
    header = income.iloc[0]
    income = income[1:]
    income.columns = header
    i = income.index[income['SA2'] == 'Victoria'].tolist()[0]
    j = income.index[income['SA2'] == 'Queensland'].tolist()[0]
    income = income.iloc[i-6:j-7, :]
    return income
```

```python
income1519 = vic_income('income_2015_2019')
income1218 = vic_income('income_2012_2018')
```
