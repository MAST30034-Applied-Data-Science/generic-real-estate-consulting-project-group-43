```python
import pandas as pd
```

```python
gdp = pd.read_csv('../../data/curated/gdp.csv', index_col=0)
gdp = gdp.reset_index()
gdp.rename({'GDP_current_prices_Millions': 'gdp(USD Millioins)'}, axis=1, inplace=True)
gdp
```

```python
# sasving rate: https://data.oecd.org/natincome/saving-rate.htm
# interest rate: https://data.oecd.org/interest/long-term-interest-rates.htm#indicator-chart
# inflation: https://data.oecd.org/price/inflation-cpi.htm
# consumer price: https://www.oecd-ilibrary.org/economics/data/prices/consumer-prices-complete-database_0f2e8000-en
saving_rate = pd.read_csv('../../data/raw/external_dataset/DP_LIVE_12092022072122298.csv')
saving_rate = saving_rate.loc[saving_rate['LOCATION']=='AUS']
saving_rate.rename({'Value': 'saving_rate(% of GDP)'}, axis=1, inplace=True)
saving_rate
```

```python
COL = ['TIME', 'saving_rate(% of GDP)']
final = gdp.merge(saving_rate[COL], left_on=['year'], right_on=['TIME']).drop(columns=['index', 'TIME'])
final = final[["year", "gdp(USD Millioins)", "saving_rate(% of GDP)"]]
final
```

```python
final.to_csv('../../data/curated/gdp_with_saving_rate.csv')
```
