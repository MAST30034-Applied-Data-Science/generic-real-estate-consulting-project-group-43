```python
import geopandas as gpd
import folium
import pandas as pd
```

```python
boundary = gpd.read_file("../../data/raw/Geo/SA2_2021_AUST_SHP_GDA2020/SA2_2021_AUST_GDA2020.shp")
boundary = boundary[boundary.STE_NAME21 == 'Victoria']
geo = boundary[['SA2_CODE21','SA2_NAME21', 'geometry']]
geo.SA2_CODE21 = geo.SA2_CODE21.astype(int)
```

```python
population_density  = pd.read_csv('../../data/curated/vic_population_density_sa2.csv', index_col = 0)
population_density['SA2 code'] = population_density['SA2 code'].astype(int)
population_density
```

```python
gdf = population_density.merge(geo, how='left', left_on='SA2 code', right_on = 'SA2_CODE21')
gdf
```
