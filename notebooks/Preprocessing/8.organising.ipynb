# Organising feature names and orders

```python
import pandas as pd
import os
```

Create path

```python
new_path = '../../data/curated/min_distance_sa2_organised/'

if not os.path.exists(new_path):
    os.makedirs(new_path)
```

Read csv files

```python
path = '../../data/curated/min_distance_sa2/'
filename = '_property_with_SA2.csv'
COL = ['year', 'month', 'address', 'latitude', 'longitude', 'postcode', 'sa2_2021', 'sa2_2016', 
        'residence_type', 'nbed', 'nbath', 'ncar', 'min_distance_to_cbd', 
        'min_distance_to_park', 'min_distance_to_prim', 'min_distance_to_second', 
        'min_distance_to_train', 'min_distance_to_hosp', 'min_distance_to_poli', 
        'min_distance_to_shop', 'weekly_rent']

for y in range(2013, 2023):
    rent = pd.read_csv(path+str(y)+filename, index_col=0)
    rent = rent.rename({'latitude_ori': 'latitude', 'longitude_ori': 'longitude', 'SA2_CODE': 'sa2_2021', 'SA2_CODE_2016': 'sa2_2016', 'dist_to_cbd_KM': 'min_distance_to_cbd'}, axis=1)
    rent[COL].to_csv(new_path+str(y)+filename)
```
